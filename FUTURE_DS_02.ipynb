```python
!pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "/content/drive/MyDrive/data.csv"
df = pd.read_csv("/content/drive/MyDrive/data.csv")

print("First 5 rows:")
display(df.head())
print("\nDataset info:")
df.info()

print("\nSummary Statistics:")
print(df.describe())

total_impressions = df['Impressions'].sum() if 'Impressions' in df.columns else 0
total_clicks = df['Clicks'].sum() if 'Clicks' in df.columns else 0
total_conversions = df['Conversions'].sum() if 'Conversions' in df.columns else 0

spend_col = next((col for col in df.columns if "Amount Spent" in col), None)
total_spent = df[spend_col].sum() if spend_col else 0

print(f"\nTotal Impressions: {total_impressions}")
print(f"Total Clicks: {total_clicks}")
print(f"Total Conversions: {total_conversions}")
print(f"Total Spend: ₹{total_spent}")

engagement_cols = [col for col in ['Likes', 'Comments', 'Shares'] if col in df.columns]
if engagement_cols:
    for col in engagement_cols:
        print(f"Total {col}: {df[col].sum()}")
else:
    print("\nEngagement data not available.")

if 'Impressions' in df.columns and 'Clicks' in df.columns:
    df['CTR (%)'] = (df['Clicks'] / df['Impressions'].replace(0, pd.NA)) * 100
    print("\nAverage CTR (%):", round(df['CTR (%)'].mean(skipna=True), 2))
else:
    print("\nCTR cannot be calculated (missing columns).")

if 'Revenue' in df.columns and spend_col:
    df['ROI (%)'] = ((df['Revenue'] - df[spend_col]) / df[spend_col].replace(0, pd.NA)) * 100
    print("\nAverage ROI (%):", round(df['ROI (%)'].mean(skipna=True), 2))
else:
    print("\nRevenue data not available for ROI calculation.")

sns.set(style="whitegrid")

if 'Campaign Name' in df.columns and 'CTR (%)' in df.columns:
    plt.figure(figsize=(10,5))
    sns.barplot(x='Campaign Name', y='CTR (%)', data=df)
    plt.xticks(rotation=45)
    plt.title('CTR by Campaign')
    plt.show()

if 'Revenue' in df.columns and spend_col:
    plt.figure(figsize=(6,6))
    sns.scatterplot(x=spend_col, y='Revenue', data=df)
    plt.title('Spend vs Revenue')
    plt.show()

if engagement_cols:
    engagement_df = df[engagement_cols].sum().reset_index()
    engagement_df.columns = ['Metric', 'Count']
    plt.figure(figsize=(6,6))
    sns.barplot(x='Metric', y='Count', data=engagement_df)
    plt.title('Total Engagement by Type')
    plt.show()
```
